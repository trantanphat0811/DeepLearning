# LAB 1: THAO TÁC DỮ LIỆU
## Người thực hiện: Trần Tấn Phát
## Xử lý dữ liệu từ file: processed_dulieuxettuyendaihoc.csv

## Bước 0: Bước chuẩn bị - Khai báo thư viện và đọc dữ liệu

```python
import pandas as pd
import numpy as np

# Đọc file dữ liệu
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

print("Dữ liệu được đọc thành công!")
print(f"Kích thước dữ liệu: {df.shape}")
print("\nHàng đầu tiên:")
print(df.head())
```

## Bước 1: Phân loại dữ liệu (Định tính & Định lượng)

```python
# Xem kiểu dữ liệu của các cột
print("=== THÔNG TIN VỀ KIỂU DỮ LIỆU ===")
print(df.info())

print("\n=== LIỆT KÊ CÁC CỘT ĐỊNH LƯỢNG (Quantitative) ===")
quantitative_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"Số lượng cột định lượng: {len(quantitative_cols)}")
print("Các cột:", quantitative_cols)

print("\n=== LIỆT KÊ CÁC CỘT ĐỊNH TÍNH (Qualitative) ===")
qualitative_cols = df.select_dtypes(include=['object']).columns.tolist()
print(f"Số lượng cột định tính: {len(qualitative_cols)}")
print("Các cột:", qualitative_cols)
```

## Bước 2 & 3: Lập bảng tần số, tần suất để khảo sát dữ liệu thiếu

```python
# Kiểm tra số lượng giá trị thiếu ở mỗi cột
print("=== SỐ LƯỢNG DỮ LIỆU THIẾU ===")
missing_data = df.isnull().sum()
missing_data_nonzero = missing_data[missing_data > 0]
if len(missing_data_nonzero) > 0:
    print(missing_data_nonzero)
else:
    print("Không có dữ liệu thiếu!")

# Tính phần trăm dữ liệu thiếu (tần suất)
print("\n=== TẦN SUẤT DỮ LIỆU THIẾU (%) ===")
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_percentage_nonzero = missing_percentage[missing_percentage > 0]
if len(missing_percentage_nonzero) > 0:
    print(missing_percentage_nonzero)
else:
    print("Không có dữ liệu thiếu!")

print(f"\n=== TỔNG SỐ DỮ LIỆU THIẾU ===")
total_missing = df.isnull().sum().sum()
print(f"Tổng cộng: {total_missing}")
```

## Bước 4: Hiệu chỉnh thang đo và kiểu giá trị

```python
# Chuyển đổi các cột phân loại thành kiểu category
categorical_cols = ['GT', 'DT', 'KV', 'KT', 'XL1', 'XL2', 'XL3']

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')

print("=== ĐÃ CHUYỂN ĐỔI KIỂU DỮ LIỆU ===")
print(f"Các cột phân loại: {categorical_cols}")
print("\nKiểu dữ liệu sau chuyển đổi:")
print(df[categorical_cols].dtypes)
```

## Bước 5: Xử lý dữ liệu thiếu bằng phương pháp Mean

```python
# Xử lý tự động cho tất cả các cột số
for col in df.select_dtypes(include=[np.number]).columns:
    if df[col].isnull().sum() > 0:
        mean_value = df[col].mean()
        df[col].fillna(mean_value, inplace=True)
        print(f"Cột {col}: Đã điền {df[col].isnull().sum()} giá trị trung bình")

print("\n=== XỬ LÝ XONG DỮ LIỆU THIẾU ===")
print(f"Tổng số dữ liệu thiếu còn lại: {df.isnull().sum().sum()}")
print("Tất cả dữ liệu đã được xử lý!")
```

## Bước 6: Thực hiện xử lý lần lượt cho các biến điểm số còn lại

```python
# Danh sách các biến điểm số từ lớp 10 (1) đến lớp 12 (6)
subjects = ['T', 'L', 'H', 'S', 'V', 'X', 'D', 'N']
years = ['1', '2', '3', '4', '5', '6']

all_score_cols = [f"{sub}{yr}" for yr in years for sub in subjects]

print("=== XỬ LÝ CÁC CỘT ĐIỂM SỐ ===")
print(f"Tổng số cột điểm số: {len(all_score_cols)}")

# Điền Mean cho tất cả các cột điểm này nếu còn thiếu
for col in all_score_cols:
    if col in df.columns:
        if df[col].isnull().sum() > 0:
            mean_val = df[col].mean()
            df[col].fillna(mean_val, inplace=True)
            print(f"Cột {col}: Đã xử lý {df[col].isnull().sum()} giá trị thiếu")

print(f"\nTổng dữ liệu thiếu còn lại: {df.isnull().sum().sum()}")
print("Đã xử lý xong tất cả các cột điểm!")
```

## Bước 7: Tính điểm trung bình môn (TBM1, TBM2, TBM3)
### Công thức: TBM = (Toán×2 + Văn×2 + Lý + Hóa + Sinh + Sử + Địa + Ngoại ngữ) / 10

```python
# Tính TBM cho năm lớp 10 (Sử dụng các cột T1, L1, H1, S1, V1, X1, D1, N1)
# T: Toán (hệ số 2)
# L: Lý
# H: Hóa
# S: Sinh
# V: Văn (hệ số 2)
# X: Sử
# D: Địa
# N: Ngoại ngữ

# Lưu ý: Dữ liệu gốc có các cột T1-N1 là lớp 10 (năm thứ 1)
#        T2-N2 là lớp 11 (năm thứ 2) 
#        T5-N5 hoặc T6-N6 là lớp 12 (năm thứ 3)

print("=== TÍNH ĐIỂM TRUNG BÌNH MÔN (TBM) ===")

# Tính TBM1 cho năm lớp 10 (Năm 1)
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10

# Tính TBM2 cho năm lớp 11 (Năm 2)
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10

# Tính TBM3 cho năm lớp 12 (Năm 3)
# Lưu ý: Trong file dữ liệu, cột năm thứ 3 của lớp 12 là T6, L6, H6... (không phải T5)
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10

print("\nKết quả TBM các năm:")
print(df[['STT', 'TBM1', 'TBM2', 'TBM3']].head(10))
print("\nThống kê TBM:")
print(df[['TBM1', 'TBM2', 'TBM3']].describe())
```

## Bước 8: Tạo biến xếp loại (XL1, XL2, XL3)
### Tiêu chí xếp loại:
- **Yếu (Y)**: TBM < 5.0
- **Trung bình (TB)**: 5.0 ≤ TBM < 6.5
- **Khá (K)**: 6.5 ≤ TBM < 8.0
- **Giỏi (G)**: 8.0 ≤ TBM < 9.0
- **Xuất sắc (XS)**: TBM ≥ 9.0

```python
def xep_loai(diem):
    """
    Hàm xếp loại học tập dựa trên điểm trung bình
    """
    if diem < 5.0:
        return 'Y'  # Yếu
    elif diem < 6.5:
        return 'TB'  # Trung bình
    elif diem < 8.0:
        return 'K'  # Khá
    elif diem < 9.0:
        return 'G'  # Giỏi
    else:
        return 'XS'  # Xuất sắc

# Áp dụng hàm cho từng năm
print("=== TẠO BIẾN XẾP LOẠI ===")

df['XL1'] = df['TBM1'].apply(xep_loai)
df['XL2'] = df['TBM2'].apply(xep_loai)
df['XL3'] = df['TBM3'].apply(xep_loai)

print("\nKết quả xếp loại các năm:")
print(df[['STT', 'TBM1', 'XL1', 'TBM2', 'XL2', 'TBM3', 'XL3']].head(15))

print("\n=== PHÂN BỐ XẾP LOẠI ===")
print("\nXếp loại năm 1 (XL1):")
print(df['XL1'].value_counts().sort_index())
print("\nXếp loại năm 2 (XL2):")
print(df['XL2'].value_counts().sort_index())
print("\nXếp loại năm 3 (XL3):")
print(df['XL3'].value_counts().sort_index())
```

## Bước 9: Chuyển đổi sang thang điểm US (GPA 4.0)
### Công thức quy đổi: (Điểm hệ 10 / 10) × 4

```python
# Công thức: (Điểm hệ 10 / 10) * 4
print("=== CHUYỂN ĐỔI SANG THANG ĐIỂM US (GPA 4.0) ===")

df['US_TBM1'] = (df['TBM1'] / 10) * 4
df['US_TBM2'] = (df['TBM2'] / 10) * 4
df['US_TBM3'] = (df['TBM3'] / 10) * 4

print("\nKết quả quy đổi GPA:")
print(df[['STT', 'TBM1', 'US_TBM1', 'TBM2', 'US_TBM2', 'TBM3', 'US_TBM3']].head(15))

print("\nThống kê GPA:")
print(df[['US_TBM1', 'US_TBM2', 'US_TBM3']].describe())
```

## Bước 10: Tạo biến Kết quả xét tuyển (KQXT)
### Tiêu chí: 
- **1.0 (Đỗ)**: Điểm trung bình thi đại học ≥ 5.0
- **0.0 (Trượt)**: Điểm trung bình thi đại học < 5.0

```python
print("=== TẠO BIẾN KẾT QUẢ XÉT TUYỂN (KQXT) ===")

# Tính điểm trung bình thi đại học
df['DIEM_DH'] = (df['DH1'] + df['DH2'] + df['DH3']) / 3

# KQXT: 1.0 là đỗ (>= 5.0), 0.0 là trượt (< 5.0)
df['KQXT'] = df['DIEM_DH'].apply(lambda x: 1.0 if x >= 5.0 else 0.0)

print("\nKết quả xét tuyển:")
print(df[['STT', 'DH1', 'DH2', 'DH3', 'DIEM_DH', 'KQXT']].head(20))

print("\n=== THỐNG KÊ KẾT QUẢ XÉT TUYỂN ===")
kqxt_counts = df['KQXT'].value_counts()
print(f"Số học sinh đỗ (1.0): {kqxt_counts.get(1.0, 0)}")
print(f"Số học sinh trượt (0.0): {kqxt_counts.get(0.0, 0)}")
print(f"Tỷ lệ đỗ: {(kqxt_counts.get(1.0, 0) / len(df) * 100):.2f}%")
```

## Bước 11: Lưu kết quả xử lý vào file mới

```python
# Lưu kết quả ra file mới
print("=== LƯU KẾT QUẢ ===")

output_file = 'ket_qua_lab1.csv'
df.to_csv(output_file, index=False)

print(f"Đã lưu file {output_file} thành công!")
print(f"\nTổng số hàng: {len(df)}")
print(f"Tổng số cột: {len(df.columns)}")
print(f"\nDanh sách các cột trong file output:")
print(df.columns.tolist())

print("\n=== KIỂM TRA LẠI DỮ LIỆU ===")
print(f"Tổng dữ liệu thiếu: {df.isnull().sum().sum()}")
print("\nSample dữ liệu cuối cùng (5 hàng cuối):")
print(df[['STT', 'GT', 'KV', 'TBM1', 'XL1', 'US_TBM1', 'TBM2', 'XL2', 'US_TBM2', 'TBM3', 'XL3', 'US_TBM3', 'KQXT']].tail())
```
